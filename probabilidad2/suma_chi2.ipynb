```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
import boost_histogram as bh
```

Número de variables a sumar

```python
numero_variables = 10
```

Número de simulaciones

```python
numero_simulaciones = 10000
```

Generador de números aleatorios

```python
rng = np.random.default_rng()
```

Histograma para guardar la suma de las variables uniformes

```python
numero_bines = 10 * numero_variables
maximo_x = 2 * numero_variables 
hist = bh.Histogram(bh.axis.Regular(bins=numero_bines, start=0, stop=maximo_x))
```

Genero los datos y lleno el histograma

```python
for i in range(numero_simulaciones):
    x = rng.standard_normal(numero_variables)
    y = np.sum(x*x)
    hist.fill(y)
```

## Grafico los datos generados

```python
fig, ax = plt.subplots()
plt.xlabel('y')
plt.ylabel('Frecuencia')
```

```python
ax.plot(hist.axes.centers[0], hist.values(), drawstyle='steps-mid', label='Simulaciones')
```

## Distribución $\chi^2$

$\chi^2_k(y) = \frac{1}{2^{k/2} \, \Gamma(k/2)}\; y^{k/2-1} e^{-y/2}\;$

Sigma

```python
print("Media = ", numero_variables)
sigma = math.sqrt(2*numero_variables)
print("Desviación estándar = ", sigma)
```

Grafico

```python
x = np.linspace(0, maximo_x, 256)
k = numero_simulaciones * hist.axes.widths[0][0]
y = k * chi2.pdf(x, df=numero_variables)
ax.plot(x, y, label='$\chi^2$')
```

```python
ax.legend(frameon=False)
```
