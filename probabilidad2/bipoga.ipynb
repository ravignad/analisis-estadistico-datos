```python
n = 400
p = 0.05
```

```python
print("Parámetros seleccionados")
print("n = ", n)
print("p = ", p)
```

Parámetro Poisson

```python
mu = n*p
print("mu = ", mu)
```

Parámetros Gauss

Parámetro μ igual a Poisson

```python
varianza = n*p*(1-p)
sigma = math.sqrt(varianza)
print("sigma = ", sigma)
```

## Grafico PDFs

```python
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
```

Binomial

```python
kmin = binom.ppf(0.01, n, p)
kmax = binom.ppf(0.99, n, p)
k = np.arange(kmin, kmax+1)
ax.plot(k, binom.pmf(k, n, p), 'v', ms=6, fillstyle='none', label='Binomial')
```

Poisson

```python
kmin = poisson.ppf(0.01, mu)
kmax = poisson.ppf(0.99, mu)
k = np.arange(kmin, kmax+1)
ax.plot(k, poisson.pmf(k, mu), '^', ms=6, fillstyle='none', label='Poisson' )
```

Normal

```python
xmax = mu + 3 * sigma
xmin = mu - 3 * sigma
x = np.linspace(xmin, xmax, 256)
ax.plot(x, norm.pdf(x, mu, sigma), label='Normal')
```

```python
ax.legend(loc='lower left')
ax.text(0.7, 0.9,  f'n = {n}', fontsize='x-small', transform=ax.transAxes)
ax.text(0.7, 0.8, f'p = {p}', fontsize='x-small', transform=ax.transAxes)
```
